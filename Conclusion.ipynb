# Concluding Remarks

**While the walkability and water feature data sets can individually decide the best neighborhood in Pittsburgh, combing these data sets leads to a more comprehensive determination of the best neighborhood in Pittsburgh. The "Combined-Data" dataset contains both the walkability and water feature values.**

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# load the walkability data directly from the WPRDC
combined = pd.read_csv("Combined-Data.csv")
combined.head(10)
```

**First, we combined the unweighted walkability and water features for each neighborhood and plotted them below. As you can see, Squirrel Hill South was given the title of best overall neighborhood in Pittsburgh while New Homestead was given the title of worst overall neighborhood in Pittsburgh.**

```python
neighborhoods = combined['Neighborhood'].values
walk = combined['Walkability'].values
water = combined['Water Features'].values
walk_w = combined['Walkability Weighted'].values
water_w = combined['Water Features Weighted'].values

plt.figure(figsize=(20, 10))
plt.bar(neighborhoods, water, width = 0.6, label='Water Features', color ='blue', bottom = walk)
plt.bar(neighborhoods, walk, width = 0.6, label='Walkability', color = 'red')

plt.xticks(neighborhoods, rotation=90)
plt.ylabel("Value")
plt.xlabel("Neighborhoods")
plt.legend(loc="upper right")
plt.title("Best Neighborhood in Pittsburgh: Walkability and Water Features")
plt.show()
```

**However, we realized that not weighting the metrics significantly impacted the data and walkability is also a more sought after quality than water features. With this in mind, we gave walkability a weight of 0.7 and water features a weight of 0.3. After calculating the weighted totals, it was determined that Allegheny Center was the best neighborhood in Pittsburgh and New Homestead remained the worst Neighborhood in Pittsburgh. Squirrel Hill South was still one of the top neighborhoods, but was no longer the best due to the change in the weight of the water feature metric.**

```python
neighborhoods = combined['Neighborhood'].values
walk_w = combined['Walkability Weighted'].values
water_w = combined['Water Features Weighted'].values

plt.figure(figsize=(20, 10))
plt.bar(neighborhoods, water_w, width = 0.6, label='Water Features', color ='purple', bottom = walk_w)
plt.bar(neighborhoods, walk_w, width = 0.6, label='Walkability', color = 'orange')

plt.xticks(neighborhoods, rotation=90)
plt.ylabel("Weighted Value")
plt.xlabel("Neighborhoods")
plt.legend(loc="upper right")
plt.title("Best Neighborhood in Pittsburgh: Walkability and Water Features Weighted")
plt.show()
```
